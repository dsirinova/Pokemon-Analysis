<h1>Type1 and Type2 Analysis</h1>

```python
df_pokemon["Type 1"].value_counts()
```

```python
df_pokemon["Type 2"].value_counts()
```

```python
df_pokemon["is_Dual"] = df_pokemon["Type 2"].notna()
dual_percentage = df_pokemon["is_Dual"].mean() * 100
dual_percentage
```

```python
df_pokemon["is_Dual"].value_counts()
```

```python
df_pokemon.groupby("Type 1")["Total"].mean().sort_values(ascending = False)
```

```python
df_pokemon.groupby("Type 2")["Total"].mean().sort_values(ascending = False)
```

```python
df_pokemon.groupby("Type 1")[["HP", "Attack", "Defense","Sp. Atk", "Sp. Def", "Speed"]].mean().round(1).sort_values(by = "HP", ascending = False)
```

```python
df_pokemon.groupby("Type 2")[["HP", "Attack", "Defense","Sp. Atk", "Sp. Def", "Speed"]].mean().round(1).sort_values(by = "HP", ascending = False)
```

```python
strongest_type1 = df_pokemon.groupby("Type 1")["Total"].mean().idxmax()
weakest_type1 = df_pokemon.groupby("Type 1")["Total"].mean().idxmin()
print(strongest_type1)
print(weakest_type1)
```

```python
df_pokemon[df_pokemon["Type 1"] == strongest_type1][["Name", "Type 1", "Total"]]
```

```python
strongest_type2 = df_pokemon.groupby("Type 2")["Total"].mean().idxmax()
weakest_type2 = df_pokemon.groupby("Type 2")["Total"].mean().idxmin()
print(strongest_type2)
print(weakest_type2)
```

```python
df_pokemon["Type_Combo"] = df_pokemon["Type 1"] + "/" + df_pokemon["Type 2"].fillna("None")
df_pokemon["Type_Combo"].value_counts()
```

```python
df_pokemon.groupby("Type_Combo")["Total"].mean().sort_values(ascending = False)
```

```python
df_pokemon.groupby("Type_Combo")["Total"].mean().idxmax()
```

```python
df_pokemon.groupby("Type_Combo")["Total"].mean().idxmin()
```

```python
strongest_name = df_pokemon.loc[df_pokemon["Total"].idxmax(), "Name"]
weakest_name = df_pokemon.loc[df_pokemon["Total"].idxmin(), "Name"]
print(strongest_name)
print(weakest_name)
```

```python
df_pokemon.iloc[df_pokemon["Total"].idxmax(), [1,4,5,6,7,8,9,10]]
```

```python
attack = df_pokemon.loc[df_pokemon["Attack"].idxmax(), ["Name", "Total"]]
defense = df_pokemon.loc[df_pokemon["Defense"].idxmax(), ["Name", "Total"]]
sp_attack = df_pokemon.loc[df_pokemon["Sp. Atk"].idxmax(), ["Name", "Total"]]
sp_defense = df_pokemon.loc[df_pokemon["Sp. Def"].idxmax(), ["Name", "Total"]]
speed = df_pokemon.loc[df_pokemon["Speed"].idxmax(), ["Name", "Total"]]
print(attack)
print(defense)
print(sp_attack)
print(sp_defense)
print(speed)
```

```python
df_pokemon.groupby("Type 1")["Speed"].mean().sort_values()
```

```python
df_pokemon.groupby("Type 2")["Speed"].mean().sort_values()
```

```python
single_mean = df_pokemon[df_pokemon["is_Dual"] == False]["Total"].mean()
dual_mean = df_pokemon[df_pokemon["is_Dual"] == True]["Total"].mean()
print("Single-type mean:", single_mean)
print("Dual-type mean:", dual_mean)
```
